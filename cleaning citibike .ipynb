```python
#read in the csv

df= pd.read_csv("Resources/2019_Citibike.csv")
df.head()
```

```python
#drop the rows where the trip duration is less than 90 seconds and where the start and end stations are the same

df =df.loc[((df["tripduration"]>90) & (df["start station id"] != df["end station id"])),:]
df.head()
```

```python
#Convert trip duration column from seconds to minutes

df["tripduration"] = [trip/60 for trip in df["tripduration"]]

#filter out trips over 24 hours long
df= df.loc[df["tripduration"] < 1440, :]

#Rename trip duration column
df.rename(columns={"tripduration": "tripduration (min)"})

df.head()
```

```python
#Rename values in gender column with strings

gender = {0: "Unknown", 1: "Male", 2: "Female"}
df.gender = [gender[item] for item in df.gender]
df.head()
```

```python
#Add ride id column to uniquely identify each ride

df["rideid"] = df["start station id"].map(str) + "_" + df["end station id"].map(str)
df.head()
```

```python
#calculate the distance for each ride

from math import sin, cos, sqrt, atan2, radians

#convert degrees to radiuns
lat1 = [radians(lat) for lat in df["start station latitude"]]
lon1 = [radians(lon) for lon in df["start station longitude"]]
lat2 =[radians(lat) for lat in df["end station latitude"]]
lon2 =[radians(lon) for lon in df["end station longitude"]]


#convert lists into series
lat1= pd.Series(lat1)
lon1= pd.Series(lon1)
lat2= pd.Series(lat2)
lon2= pd.Series(lon2)

#calculate the difference between each set of long and lat

dlat = lat2 - lat1
dlon = lon2 - lon1
```

```python
#approximate the radius of the earth in km

R= 6373.0


#empty list to store trip distances

distance= []

for i in range(0, len(lat1)):
    a = sin(dlat[i] / 2)**2 + cos(lat1[i]) * cos(lat2[i]) * sin(dlon[i] / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Get distance and convert km to miles
    miles = (R * c) * .6214
    
    # Append miles travel to 'distance' list
    distance.append(miles)
```

```python
#Add trip distance as new column to dataframe

df["tripdistance (mi)"] = distance
df.head()
```

```python
#calculate the age of the riders

ages = []

for i in df["birth year"]:
    age = 2019 - i
    ages.append(age)
```

```python
#Append new age 

df["Ages"] = ages
df
```

```python
#Delete all rows where user age is over 90

df= df.loc[df["Ages"] < 90]
```

```python
# Divide users in age bins
bins = [0, 17, 25, 35, 50, 65, 90]
labels = ["0-17", "18-25", "26-35", "36-50", "51-65", "65-90"]

df["Age Group"] = pd.cut(df["Ages"], bins, labels=labels)
```

```python
df.to_csv("Resources/2019-citibike-all.csv", index=False)
```
